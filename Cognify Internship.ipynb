```python
#Importing necessary Libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import plotly.express as px
import missingno as msno
import warnings
warnings.filterwarnings('ignore')
```

```python
#Loading the Dataset
df = pd.read_csv('Dataset.csv')
```

```python
#viewing first few rows
df.head()
```

```python
#checking dtypes
df.dtypes
```

```python
#checking any null values
df.isnull().sum()
```

```python
msno.bar(df)
plt.show()
```

```python
#Handling the missing values of Cuisines
cuisine_most_freq = df['Cuisines'].mode()[0]
df['Cuisines'].fillna(cuisine_most_freq,inplace = True)
```

```python
#checking duplicate values
df.duplicated().any()
```

## Level 1

### Task 1 - Top Cuisines

```python
#Determining top three common Cuisines
top_cuisine = df['Cuisines'].value_counts().iloc[:3]

#Calculating the percentage of restaurants that serve each of the top cuisines
top_cuisine_percent = ((top_cuisine/df['Cuisines'].size))*100

print(top_cuisine_percent)
```

### Task 2 - City Analysis

```python
#Identifying the city with the highest number of restaurants
Top_city = df['City'].value_counts().idxmax()

# Calculating the average rating for restaurants in each city
Avg_rating = df.groupby(['City'])['Aggregate rating'].mean()

#Determining the city with the highest average rating
highest_avg_rating_city = Avg_rating.idxmax()


print('City with highest number of Restaurant: ',Top_city,'\n')
print('Average rating for restaurants in each city: ',Avg_rating,'\n')
print('City with highest average rating: ',highest_avg_rating_city)
```

### Task 3: Price Range Distribution


```python
# Visualizing the distribution of price ranges among the restaurants
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Price range', order=df['Price range'].value_counts().index)
plt.title('Distribution of Price ranges Among Restaurants')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.show()

# Calculating the percentage of restaurants in each price range category
Percent_of_restaurant = ((df['Price range'].value_counts()/df['Price range'].size))*100
print('Percentage of restaurants in each price range category: ',Percent_of_restaurant)
```

### Task 4: Online Delivery

```python
# Determining the percentage of restaurants that offer online delivery
online_delivery = (df['Has Online delivery'].eq('Yes').mean())*100
print('The percentage of restaurants that offer online delivery: ',online_delivery,'\n')

#Comparing average ratings of restaurants with and without online delivery
df['Has Online delivery'] = df['Has Online delivery'].astype('category')

plt.figure(figsize=(10, 6))
sns.boxplot(x='Has Online delivery', y='Aggregate rating', data=df)
plt.title('Comparison of Aggregate Ratings for Restaurants With and Without Online Delivery')
plt.xlabel('Has Online Delivery')
plt.ylabel('Aggregate Rating')
plt.show()
```

## Level 2

### Task 1 - Restaurant Rating

```python
# Analyzing the distribution of aggregate ratings
df['Aggregate rating'].plot(kind='hist', bins=20)
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Rating')
plt.show()

#Determining the most common rating range
common_rating = df['Aggregate rating'].mode()[0]
print('Most common rating range: ',common_rating,'\n')

# Calculate the average number of votes received by restaurants
avg_vote= df['Votes'].mean()
print('The average number of votes received by restaurants: ',avg_vote)
```

### Task 2 - Cuisine Combination

```python
#Identifyng most common combinations of cuisine
most_common = df['Cuisines'].value_counts().head(3)
print('The most common combinations of cuisine: ',most_common,'\n')

# Determining if certain cuisine combinations tend to have higher ratings
cuisine_rating = df.groupby('Cuisines')['Aggregate rating'].mean()
cuisine_Higher_rating = cuisine_rating[cuisine_rating == cuisine_rating.max()]
print('Certain cuisine combinations tend to have higher ratings: ',cuisine_Higher_rating)
```

### Task 3: Geographic Analysis


```python
# Ploting the locations of restaurants on a map using longitude and latitude coordinates

plt.scatter(df['Longitude'], df['Latitude'], alpha=0.5,c='blue')
plt.title('Geographic Distribution of Restaurants')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

# Identify any patterns or clusters of restaurants in specific areas
# For clustering, we could use  KMeans
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5) #Assuming 5 clusters
df['cluster'] = kmeans.fit_predict(df[['Longitude', 'Latitude']])

#Plotting the clusters
plt.figure(figsize=(12, 8))
plt.scatter(df['Longitude'], df['Latitude'], c=df['cluster'], cmap='viridis', alpha=0.5)
plt.title('Clustered Geographic Distribution of Restaurants')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

### Task 4 - Restaurant Chains

```python
# Identifying if there are any restaurant chains in the dataset
restaurant_counts = df['Restaurant Name'].value_counts()
chains = restaurant_counts[restaurant_counts > 1]
print('Restaurant chains: ',chains,'\n')

# Ratings and Popularity different restaurant chains
chain_rating = df.groupby('Restaurant Name')['Aggregate rating'].mean().sort_values(ascending=False)
print('Ratings of Different chains: ',chain_rating,'\n')

chain_popularity = df['Restaurant Name'].value_counts()
print('Popularity of Different Restaurant: ',chain_popularity)


```
